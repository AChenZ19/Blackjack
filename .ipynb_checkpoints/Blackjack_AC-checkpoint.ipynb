#### 1. Create 52 Set of Card Objects (Face, Value)
- create_card_set(): returns 52 card set
- show_face(card_id)
- show_value(card_id)
- class: Card(face,value)

```python
# function to determine 'card face' from 'card ID'
# could use dictionary instead

import numpy as np
import math

def show_face(card_id):
    x = card_id % 13 + 2
    if x == 11:
        card_rank = "Jack"
    elif x == 12:
        card_rank = "Queen"
    elif x == 13:
        card_rank = "King"
    elif x == 14:
        card_rank = "Ace"
    else:
        card_rank = str(x)
    if card_id < 13:
        card_suit = "Clubs"
    elif card_id < 26:
        card_suit = "Diamonds"
    elif card_id < 39:
        card_suit = "Hearts"
    else:
        card_suit = "Spades"
    face = (str(card_rank) + " of " + str(card_suit))
    return face
```

```python
# function to determine 'card value' from 'card ID'
# assume Ace has card value of 11, value-corrected to logical max when evaluated

def show_value(card_id):
    x = card_id % 13 + 2
    if x < 11:
        card_value = x
    elif x < 14:
        card_value = 10
    else:
        card_value = 11
    return card_value
```

```python
# consider Hand class (attributes: bust - boolean, aces_held, ownership - for splits, etc.)
# and using methods to modify on data in each class

class Card:
    def __init__(self, face, value):
        self.face = face
        self.value = value

def create_card_set():
    set_ids = list(range(52))
    set_cards = []
    for x in set_ids:
        set_cards.append(Card(show_face(x), show_value(x)))
    return set_cards
```

#### 2. Create Dealer, Players, Scoreboard
- create_dealer()
- dealer_hand
- enter_players()
- player_hands
- num_of_players

```python
# function to create dealer and enter desired number of players

def create_dealer():
    global dealer_hand
    dealer_hand = []

def enter_players():
    global num_of_players
    num_of_players = int(input("How many players would you like to enter? "))
    global player_hands
    player_hands = [[] for _ in range(num_of_players)]
    
def create_record_sheet():
    global players_W
    global players_L
    global players_T
    players_W = [0] * num_of_players
    players_L = [0] * num_of_players
    players_T = [0] * num_of_players
```

#### 3. Create Game Deck
- create_game_deck(): returns game-ready deck
- create_full_deck()
- shuffle()
- insert_cut(target)
- cut_card

```python
# function to generate deck of desired number of sets

def create_full_deck():
    target_sets = int(input("How many sets would you like to play with? "))
    full_deck = list()
    sets = 1
    set_cards = create_card_set()
    while sets <= target_sets:
        full_deck += set_cards
        sets += 1
    return full_deck
```

```python
# function to shuffle target

def shuffle(target):
    np.random.shuffle(target)
    return target
```

```python
# function to insert 'cut card' randomly into second half of deck to prompt reshuffle 

cut_card = Card("JOKER", None)

def insert_cut(target):
    confirm_cut_card = str(input("Type 'yes' if you would like to reshuffle using the joker: "))
    if confirm_cut_card == "yes":
        x = np.random.randint(math.floor(len(target)*0.75),len(target))
        cut_target = target.insert(x, cut_card)
    return cut_target
```

```python
# function to create game deck
# (no cut_card used)

def create_game_deck():
    global game_deck
    x = create_full_deck()
    game_deck = shuffle(x)
```

#### 5. Functions to Deal Cards and Clear Table
- deal_card(target_hand)
- clear_table(): clears table into discard_pile

```python
# function to deal cards

def deal_card(target_hand):
    global game_deck
    drawn_card = game_deck.pop(0)
    target_hand.append(drawn_card)
    if target_hand == dealer_hand:
        if len(dealer_hand) == 1:
            print("A card was dealt facedown to the Dealer.")
        else:
            print(drawn_card.face + " was dealt to Dealer.")
    else:
        for a in range(num_of_players):
            if target_hand == player_hands[a]:
                print(drawn_card.face + " was dealt to Player " + str(a) + ".")
#     if drawn_card.face == "JOKER":
#         print("The JOKER has been drawn. The Deck and Discard Pile will be reshuffled.")
#         game_deck = game_deck + discard_pile
#         shuffle(game_deck)
```

```python
# function to clear table cards into discard pile

def clear_table():
    global discard_pile
    discard_pile = []
    for a in range(len(dealer_hand)):
        clear_card = dealer_hand.pop(0)
        discard_pile.append(clear_card)
    for a in range(len(player_hands)):
        for b in range(len(player_hands[a])):
            clear_card = player_hands[a].pop(0)
            discard_pile.append(clear_card)
```

#### 6. Function to Start New Round
- start_round()

```python
# function to start new round

def start_round():
    global rounds
    clear_table()
    print("\nBegin Round " + str(rounds))
    for a in range(num_of_players):
        deal_card(player_hands[a])
    deal_card(dealer_hand)
    for a in range(num_of_players):
        deal_card(player_hands[a])
    deal_card(dealer_hand)
    
```

#### 7. Player Actions
- x for hit(target_hand)
- y for stand(): activates turn_complete boolean
- z for check_table(): shows cards and hand values
- s for show_record(): show W/L/T score
- show_full(target_hand): shows full hand
- show_part(): shows Dealer's hand pre-reveal
-

```python
# function to determine hand value

def total_values(target_hand):
    hand_total = 0
    for a in range(len(target_hand)):
        hand_total = hand_total + target_hand[a].value
    for b in range(len(target_hand)):
        if hand_total > 21 and target_hand[b].face[0:3] == "Ace":
            hand_total = hand_total - 10
    return hand_total
```

```python
# function for *player* action 'hit'

def hit(target_hand):
    hand_total = total_values(target_hand)
    if hand_total <= 21:
        deal_card(target_hand)
        show_full(target_hand)
    else:
        show_full(target_hand)
        print("The player has busted and cannot take additional cards.")
```

```python
# functions to show hands

def show_full(target_hand):
    hand_total = total_values(target_hand)
    target_hand_faces = []
    i = 0
    for a in range(len(target_hand)):
        target_hand_faces.append(target_hand[i].face)
        i += 1
    if target_hand == dealer_hand:
        print("Dealer hand: " + str(target_hand_faces))
        print("Dealer score: " + str(hand_total))
    else:
        for a in range(num_of_players):
            if target_hand == player_hands[a]:
                print("Player " + str(a) + " hand: " + str(target_hand_faces))
                print("Player " + str(a) + " score: " + str(hand_total))
        
def show_part():
    print("Dealer hand: [ 'Facedown Card', '" + str(dealer_hand[1].face) + "' ]")
    print("Dealer shown score: " + str(dealer_hand[1].value))
    
```

```python
# function to show all cards on table
# fix print formatting...

def check_table():
    hand_total = total_values(dealer_hand)
    show_part()
    for x in range(num_of_players):
        total_values(player_hands[x])
        show_full(player_hands[x])
```

```python
# function to evaluate wins and losses of the round

def evaluate_round():
    global players_W
    global players_L
    global players_T
    dealer_score = total_values(dealer_hand)
    i = 0
    for x in player_hands:
        player_score = total_values(x)
        if player_score > 21:
            if dealer_score > 21:
                players_L[i] += 1
                print("Player " + str(i) + " loses by busting with a score of " + str(player_score) + " before Dealer busts with a score of " + str(dealer_score) + ".")
            elif dealer_score == 21:
                players_L[i] += 1
                print("Player " + str(i) + " loses by busting with a score of " + str(player_score) + " against Dealer's Blackjack.")
            else:
                players_L[i] += 1
                print("Player " + str(i) + " loses by busting with a score of " + str(player_score) + " a gainst Dealer's score of " + str(dealer_score) + ".")
        elif player_score == 21:
            if dealer_score > 21:
                players_W[i] += 1
                print("Player " + str(i) + " wins with Blackjack against Dealer busting with a score of " + str(dealer_score) + ".")
            elif dealer_score == 21:
                players_T[i] += 1
                print("Player " + str(i) + " ties Dealer with Blackjacks.")
            else:
                players_W[i] += 1
                print("Player " + str(i) + " wins with Blackjack against Dealer's score of " + str(dealer_score) + ".")
        else:
            if dealer_score > 21:
                players_W[i] += 1
                print("Player " + str(i) + " wins with a score of " + str(player_score) + " against Dealer busting with a score of " + str(dealer_score) + ".")
            elif dealer_score == 21:
                players_L[i] += 1
                print("Player " + str(i) + " loses with a score of " + str(player_score) + " against Dealer's Blackjack.")
            elif player_score > dealer_score:
                players_W[i] += 1
                print("Player " + str(i) + " wins with a score of " + str(player_score) + " against Dealer's score of " + str(dealer_score) + ".")
            elif player_score < dealer_score:
                players_L[i] += 1
                print("Player " + str(i) + " loses with a score of " + str(player_score) + " against Dealer's score of " + str(dealer_score) + ".")
            else:
                players_T[i] += 1
                print("Player " + str(i) + " ties with a score of " + str(player_score) + " against Dealer's score of " + str(dealer_score) + ".")
        i += 1
```

```python
# function to check full record W/L/T

def show_record():
    for a in range(num_of_players):
        print("Player " + str(a) + ": W-" + str(players_W[a]) + " L-" + str(players_L[a]) + " T-" + str(players_T[a]))
    
```

#### 8. Functions for Dealer and Player Turns
- dealer_turn()
- player_turns()

```python
# function to conduct dealer's turn

def dealer_turn():
    print("\nIt is the Dealer's turn.")
    show_full(dealer_hand)
    dealer_score = total_values(dealer_hand)
    while dealer_score < 17:
        deal_card(dealer_hand)
        show_full(dealer_hand)        
        dealer_score = total_values(dealer_hand)
```

```python
# function for players turn
# build in some error catching?

def player_turns():
    for a in range(num_of_players):
        turn_complete = False
        while turn_complete == False:
            action = input("\nWhat would Player " + str(a) + " like to do? Press: \n (x) to Hit \n (y) to Stand \n (z) to Check Table \n (s) to Show Record \n")
            if action == "x":
                hit(player_hands[a])
            elif action == "y":
                turn_complete = True
            elif action == "z":
                check_table()
            elif action == "s":
                show_record()
            else:
                print("Please select valid player action until you are ready to Stand (y).")
                
```

#### 9. Functions to Conclude Segments
- conclude_round()
- conclude_game()

```python
# function to conclude round

def conclude_round():
    print("\nRound " + str(rounds) + " has ended. \nThe scoreboard is:")
    show_record()
```

```python
# function to conclude game

def conclude_game():
    print("\n")
    show_record()
    tied_wins = False
    tied_ties = False
    for a in range(len(players_W)):
        for b in range(len(players_W)):
            if a != b and players_W[a] == players_W[b]:
                tied_wins = True
                for a in range(len(players_T)):
                    for b in range(len(players_T)):
                        if a != b and players_T[a] == players_T[b]:
                            tied_ties = True
                        else:
                            winner = players_T.index(max(players_T))
            else:
                winner = players_W.index(max(players_W))
    if tied_wins:
        print("There is a tie in the number of Wins. Go to tiebreaker.")        
        if tied_ties:
            print("There is a tie in the number of Ties.\nThe Champion is inconclusive.")
        else:
            print("Winner winner, chicken dinner! The Champion is Player " + str(winner) + " by tiebreaker!")
    else:
        print("Winner winner, chicken dinner! The Champion is Player " + str(winner) + "!")
```

#### 10. Function to Play a Game of Blackjack
- play_blackjack()

```python
# function to facilitate full game
# manually feed scanerios to test all cases...

def play_blackjack():
    global rounds
    rounds = 1
    target_rounds = input("How many rounds would you like to play? ")
    init_new_game()
    while rounds <= int(target_rounds):
        start_round()
        player_turns()
        dealer_turn()
        evaluate_round()
        conclude_round()
        rounds += 1
    conclude_game()
```
